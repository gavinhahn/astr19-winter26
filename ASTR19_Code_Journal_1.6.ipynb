Create a Jupyter Notebook where, in separate cells, you define functions that return sin(x) and cos(x). 
Use Markdown cells to comment your Notebook, and describe what each function does. Create a third Python 
cell that will tabulate sin(x) and cos(x) using these previously defined functions vs. x, where x is 
tabulated between 0 and 2 with a thousand entries. Write a fourth Python cell that will use a for loop 
to print out the first 10 values of x, sin(x), and cos(x) in columns.

```python
import numpy as np
```

Sin(x) Function

```python
def sin(x):
    return np.sin(x)
```

Cos(x) Function

```python
def cos(x):
    return np.cos(x)
    
```

Tabulate Function

Create sin(x) and cos(x) columns. Compile x, sin(x), and cos(x) into an array. Formatted for 3 decimals.

```python
def tabulate_data(x):
    sinx = sin(x)
    cosx = cos(x)
    
    table_data = [
        [f"{a:.3f}", f"{b:.3f}", f"{c:.3f}"]
        for a, b, c in zip(x, sinx, cosx)]
    
    return table_data
```

Main (Print 10 rows of x, sin(x), cos(x))

Uses np.linspace to create x array. Calls tabulate_data(). Prints a header. For loop to print first 10 rows.

```python
x = np.linspace(0,2*np.pi,1000)
table_data = tabulate_data(x)

print("+-------+----------+----------+")
print("|     x |   sin(x) |   cos(x) |")
print("+-------+----------+----------+")

for i in range(10):
    j = table_data[i]
    print(f"|{j[0]:>6} |{j[1]:>9} |{j[2]:>9} |")
    print("+-------+----------+----------+")
        
        
```
